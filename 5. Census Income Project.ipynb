```python
# Find the IQR to identify the outliers

# 1st quantile
q1 = data.quantile(0.25)

# 3rd quantile
q3 = data.quantile(0.75)

#IQR
iqr = q3 - q1
iqr
```

```python
# validating one outlier
cap_gain = q3.Capital_gain + (1.5*iqr.Capital_gain)
index = np.where(data['Capital_gain']>cap_gain)
data = data.drop(data.index[index])
data.reset_index().head()
```

```python
cap_loss = q3.Capital_loss + (1.5*iqr.Capital_loss)
index1 = np.where(data['Capital_loss']>cap_loss)
data = data.drop(data.index[index1])
data.reset_index().head()
```

### Since our target data is imbalanced, we have to deal with it first

### Model building for prediction of Income

```python
x = data.drop('Income',axis = 1)
y = data.Income

dt = DecisionTreeClassifier()
kn = KNeighborsClassifier()
bg = BaggingClassifier()
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
xgb = xgb.XGBRegressor()
svc = SVC()
svr = SVR()
```

```python
for i in range(0,20):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = i)
    gb.fit(x_train,y_train)
    y_train_pred = gb.predict(x_train)
    y_test_pred = gb.predict(x_test)
    print(f"At random state {i}, the training accuracy is : {accuracy_score(y_train,y_train_pred)}")
    print(f"At random state {i}, the testing accuracy is : {accuracy_score(y_test,y_test_pred)}\n")
```

### Cross Validation

```python
test_acc = accuracy_score(y_test,y_test_pred)

for i in range (2,10):
    cv_score = cross_val_score(gb,x,y,cv = i)
    cv_mean  = cv_score.mean()
    print(f"At cv = {i},cv_score = {cv_mean} and accuracy for testing = {test_acc}\n")
```
